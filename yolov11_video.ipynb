```python
# üé¨ YOLOv11 Object Detection (Colab Í≤ΩÎüâ Î≤ÑÏ†Ñ)
# Ïú†ÌäúÎ∏å ÏòÅÏÉÅ 15Ï¥àÎßå, Ìï¥ÏÉÅÎèÑ 480pÎ°ú Ï§ÑÏó¨ÏÑú YOLO ÌÉêÏßÄ ÌõÑ Ïû¨ÏÉù

!pip install -q --upgrade yt-dlp ultralytics ffmpeg-python

from ultralytics import YOLO
from IPython.display import Video, display
import os, glob, ffmpeg

# ======================================
# 1Ô∏è‚É£ Ïú†ÌäúÎ∏å ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú (15Ï¥àÎßå)
# ======================================
VIDEO_URL = "https://www.youtube.com/shorts/sEdxVrxoAc8"
os.makedirs("videos", exist_ok=True)
RAW_VIDEO = "videos/raw.mp4"
RESIZED_VIDEO = "videos/input_480p.mp4"

# 0~15Ï¥àÎßå Îã§Ïö¥Î°úÎìú
!yt-dlp -f "mp4" -o "{RAW_VIDEO}" --download-sections "*0-15" {VIDEO_URL} || echo "‚ö†Ô∏è Îã§Ïö¥Î°úÎìú Ïã§Ìå®"

print("\nüìÅ Îã§Ïö¥Î°úÎìúÎêú ÌååÏùº:")
!ls videos

# ======================================
# 2Ô∏è‚É£ Ìï¥ÏÉÅÎèÑ Ï§ÑÏù¥Í∏∞ (FFmpeg 480p Î¶¨ÏÇ¨Ïù¥Ï¶à)
# ======================================
print("\nüéûÔ∏è ÏòÅÏÉÅ Ìï¥ÏÉÅÎèÑ 480pÎ°ú Î≥ÄÌôò Ï§ë...")
ffmpeg.input(RAW_VIDEO).output(RESIZED_VIDEO, vf='scale=-2:480', vcodec='libx264', acodec='aac', strict='experimental').run(overwrite_output=True)

print("\n‚úÖ Î¶¨ÏÇ¨Ïù¥Ï¶à ÏôÑÎ£å!")

# ======================================
# 3Ô∏è‚É£ YOLOv11 ÌÉêÏßÄ ÏàòÌñâ
# ======================================
model = YOLO("yolo11n.pt")  # YOLOv11 Nano (Í≤ΩÎüâ)
results = model.predict(source=RESIZED_VIDEO, save=True, conf=0.25)

# ======================================
# 4Ô∏è‚É£ Í≤∞Í≥º ÏòÅÏÉÅ Í≤ΩÎ°ú ÌÉêÏÉâ
# ======================================
output_dir = results[0].save_dir
output_videos = glob.glob(os.path.join(output_dir, "*.mp4")) + glob.glob(os.path.join(output_dir, "*.avi"))

if not output_videos:
    print("‚ö†Ô∏è Í≤∞Í≥º ÏòÅÏÉÅÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")
else:
    raw_output = output_videos[0]
    fixed_output = os.path.join(output_dir, "output_fixed.mp4")

    print("\nüéûÔ∏è ÏµúÏ¢Ö Ïû¨Ïù∏ÏΩîÎî© (Colab Ïû¨ÏÉù Ìò∏Ìôò)")
    ffmpeg.input(raw_output).output(fixed_output, vcodec='libx264', acodec='aac', strict='experimental').run(overwrite_output=True)

    print(f"\n‚úÖ ÌÉêÏßÄ ÏôÑÎ£å! Í≤∞Í≥º ÌååÏùº: {fixed_output}")
    display(Video(fixed_output, embed=True))
```
